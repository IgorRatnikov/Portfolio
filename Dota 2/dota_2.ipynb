#### Проверка на наличие пропусков

```python
def check_missing_values(df, df_name):
    """
    Функция для подсчета и отображения пропущенных значений в столбцах датафрейма.
    """
    # Подсчет пропусков по колонкам
    report = df.isna().sum()

    # Фильтрация только колонок с пропусками
    report = report[report > 0].to_frame(name='missing_values')

    # Проверка наличия пропусков и создание отчета
    if not report.empty:
        # Добавление столбца с процентами пропусков
        report['% of total'] = (report['missing_values'] / df.shape[0]).round(2)
        report = report.sort_values(by='missing_values', ascending=False)
        display(report)
    else:
        print(f"Пропусков в таблице {df_name} нет.")

check_missing_values(heroes, 'heroes')
check_missing_values(players, 'players')
check_missing_values(matches, 'matches')
```

 - **Урон по башням в минуту**: Корреляция (0.4) указывает, что команды, наносящие больше урона по башням, чаще побеждают.
 - **Опыт в минуту** и **Золото в минуту**: Корреляции (0.38 и 0.34) показывают, что больше опыта и золота в минуту повышают шансы на победу.
 - **Убийства в минуту** и **Убийства**: Корреляции (0.35 и 0.33) указывают на то, что агрессивная стратегия с большим количеством убийств способствует победе.
 - **Урон героя в минуту**: Корреляция (0.21) показывает, что урон героя в минуту менее значим для победы по сравнению с другими факторами.

#### Среднее количество золота за минуту в зависимости от фракции (Свет/Тьма)

```python
# Создание нового столбца с более понятными названиями для фракций
players['faction'] = players['isRadiant'].map({True: 'Свет', False: 'Тьма'})

# Настройка стиля и палитры
sns.set_style('whitegrid')
sns.set_palette('pastel')

# Визуализация
plt.figure(figsize=(8, 6))
sns.boxplot(data=players, x='faction', y='gold_per_min', hue='faction', palette='Set2')
plt.title('Среднее количество золота за минуту в зависимости от фракции', fontsize=14)
plt.xlabel('Фракция', fontsize=12)
plt.ylabel('Золото за минуту', fontsize=12)

# Показываем график
plt.show()

# Расчет среднего количества золота за минуту для каждой фракции
mean_gold_per_min = round(players.groupby('isRadiant')['gold_per_min'].mean(), 1)
print(mean_gold_per_min)
```

#### Длительность матча в зависимости от фракции (Свет/Тьма)

```python
plt.figure(figsize=(14, 6))
sns.histplot(matches[matches['radiant_win'] == 1]['duration'], bins=60, kde=True, label='Победа Света')
sns.histplot(matches[matches['radiant_win'] == 0]['duration'], bins=60, kde=True, label='Победа Тьмы')
plt.legend()
plt.title('Распределение длительности матча в зависимости от победы фракции')
plt.xlabel('Длительность матча (в минутах)')
plt.ylabel('Количество побед')
plt.show()
```

```python
# Вычисляем среднее значение длительности матчей для победы Света и Тьмы
mean_duration_radiant_win = matches[matches['radiant_win'] == 1]['duration'].mean()
mean_duration_dire_win = matches[matches['radiant_win'] == 0]['duration'].mean()

print(f"Средняя длительность матча при победе Света: {mean_duration_radiant_win:.2f} минут")
print(f"Средняя длительность матча при победе Тьмы: {mean_duration_dire_win:.2f} минут")

# Вычисляем разницу в средних значениях
difference = mean_duration_dire_win - mean_duration_radiant_win
print(f"Разница в средней длительности матчей: {difference:.2f} минут")
```

Обе фракции имеют почти одинаковое распределение длительности матча в зависимости от победы. Разница лишь в том, что у Света большее количество побед в зависимости от длительности матча.

За тьму сложнее играть, поэтому времени для победы требуется в среднем чуть больше времени (свет быстрее выигрывает).

### 4. Проверка гипотез

**Герои, наносящие больше урона в минуту (Hero Damage per minute), чаще приводят к победе.**
   - **Нулевая гипотеза (H0)**: Средний урон в минуту не отличается у победивших и проигравших команд.
   - **Альтернативная гипотеза (H1)**: Средний урон в минуту выше у победивших команд.
   - **Уровень значимости (alpha)**: 0.05.
   - **Метод проверки гипотез**: T-тест.

```python
# Группировка данных по победителям и проигравшим
winners_damage = players[players['win'] == 1]['hero_damage_per_min']
losers_damage = players[players['win'] == 0]['hero_damage_per_min']

# Дополнительная информация о выборках
print(f"Размер выборки победителей: {len(winners_damage)}")
print(f"Размер выборки проигравших: {len(losers_damage)}")
print("")
print(f"Средний урон в минуту у победителей: {np.mean(winners_damage):.2f}")
print(f"Средний урон в минуту у проигравших: {np.mean(losers_damage):.2f}")
print("")

# Проведение t-теста с учетом неравенства дисперсий (тест Уэлча)
t_test_result = stats.ttest_ind(winners_damage, losers_damage, equal_var=False)
print(f'Результат t-теста (тест Уэлча): {t_test_result}')

# Интерпретация результатов
alpha = 0.05
if t_test_result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу: Средний урон в минуту значимо выше у победивших команд.")
else:
    print("Не удалось отвергнуть нулевую гипотезу: Средний урон в минуту не отличается у победивших и проигравших команд.")
```

**Игроки, набравшие больше опыта за минуту (XP per minute), с большей вероятностью выигрывают матч.**
   - **Нулевая гипотеза (H0)**: Средний XP per minute не отличается у победителей и проигравших команд.
   - **Альтернативная гипотеза (H1)**: Средний XP per minute выше у победителей.
   - **Уровень значимости (alpha):** 0.05.
   - **Метод проверки гипотез:** T-тест.

```python
# Разделение на две группы: победители и проигравшие
winners = players[players['win'] == 1]['xp_per_min']
losers = players[players['win'] == 0]['xp_per_min']

# Дополнительная информация о выборках
print(f"Размер выборки победителей: {len(winners)}")
print(f"Размер выборки проигравших: {len(losers)}")
print("")
print(f"Среднее количество опыта в минуту у победителей: {np.mean(winners):.2f}")
print(f"Среднее количество опыта в минуту у проигравших: {np.mean(losers):.2f}")
print("")

# Проведение T-теста
t_stat, p_value = ttest_ind(winners, losers, equal_var=False)  # Используем T-тест для независимых выборок

# Вывод результатов
print(f"T-статистика: {t_stat:.4f}")
print(f"P-значение: {p_value:.10f}")

# Проверка гипотезы
alpha = 0.05
if p_value < alpha:
    print("Отклоняем нулевую гипотезу. Средний XP per minute у победителей значительно выше, чем у проигравших.")
else:
    print("Не удалось отклонить нулевую гипотезу. Нет статистически значимых различий в среднем XP per minute между победителями и проигравшими.")
```

**Герои с высокой скоростью передвижения (move speed) чаще приводят к победе.**
   - **Нулевая гипотеза (H0):** Средняя скорость передвижения героев не отличается у победивших и проигравших команд.
   - **Альтернативная гипотеза (H1):** Средняя скорость передвижения героев выше у победивших команд.
   - **Уровень значимости (alpha):** 0.05.
   - **Метод проверки гипотез:** T-тест.

```python
# Объединение данных о героях и игроках по идентификатору героя
players_heroes = pd.merge(players, heroes[['hero_id', 'hero_name', 'move_speed']], left_on='hero_id', right_on='hero_id')

# Разделение данных на победителей и проигравших
winners = players_heroes[players_heroes['win'] == 1]
losers = players_heroes[players_heroes['win'] == 0]

# Средняя скорость передвижения для победивших и проигравших команд
mean_move_speed_winners = winners['move_speed'].mean()
mean_move_speed_losers = losers['move_speed'].mean()

print(f'Средняя скорость передвижения героев у победивших команд: {mean_move_speed_winners:.2f}')
print(f'Средняя скорость передвижения героев у проигравших команд: {mean_move_speed_losers:.2f}')

# Проведение T-теста
t_stat, p_value = ttest_ind(winners['move_speed'], losers['move_speed'], equal_var=False)

# Уровень значимости
alpha = 0.05

print(f"T-статистика: {t_stat:.4f}")
print(f"P-значение: {p_value:.4f}")

# Проверка результатов гипотезы
if p_value < alpha:
    print("Отвергаем нулевую гипотезу: Средняя скорость передвижения героев выше у победивших команд.")
else:
    print("Не удалось отвергнуть нулевую гипотезу: Средняя скорость передвижения героев не отличается у победивших и проигравших команд.")
```
